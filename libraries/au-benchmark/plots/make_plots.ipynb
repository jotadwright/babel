```python
def percentage_increase(orig_value, new_value, min_value):
    if orig_value > min_value and new_value > min_value:
        difference = new_value - orig_value
        increase = difference / orig_value
        percentage = increase * 100
        return percentage
```

```python
def percentage_of(value, base_value):
    if value > -1 and base_value > -1:
        frac = value / base_value
        percentage = frac * 100
        return percentage
```
