```python
# Because I always forget how to do this.
#
# Credit: https://scipython.com/blog/visualizing-the-bivariate-gaussian-distribution/

import numpy as np
import matplotlib.pyplot as plt
from   matplotlib import cm
from   mpl_toolkits.mplot3d import Axes3D
from   scipy.stats import multivariate_normal

# Our 2-dimensional distribution will be over variables X and Y
N = 60
X = np.linspace(0, 1, N)
Y = np.linspace(0, 1, N)
X, Y = np.meshgrid(X, Y)

# Mean vector and covariance matrix
#mu = np.array([0., 1.])
#Sigma = np.array([[ 1. , -0.5], [-0.5,  1.5]])

mu = np.array([0.3, 0.7])
sigma = np.array([0.01, 0.01])
Sigma = np.diag(sigma)

# Pack X and Y into a single 3-dimensional array
pos = np.empty(X.shape + (2,))
pos[:, :, 0] = X
pos[:, :, 1] = Y

# The distribution on the variables X, Y packed into pos.
F = multivariate_normal(mu, Sigma)
Z = F.pdf(pos)

# Create a surface plot and projected filled contour plot under it.
fig = plt.figure()
ax = fig.add_subplot(projection = "3d")
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, linewidth=1, 
                cmap=cm.viridis, )

# cset = ax.contourf(X, Y, Z, zdir='z', offset=-0.15, cmap=cm.viridis)

# Adjust the limits, ticks and view angle
#ax.set_zlim(0,1)
#ax.set_zticks(np.linspace(0,1.0,5))
# ax.view_init(21, -21)

plt.show()
```

```python
# x = np.array([0.3, 0.1])
mu = np.array([0.3, 0.4])
sigma = np.array([0.13, 0.001])

# make sigma into a diagonal matrix
sigma = np.diag(sigma)

# calculate the pdf
rv = multivariate_normal(mean=mu, cov=sigma)

# calculcate pdf at x
x = np.array([0.311, 0.421])
p = rv.pdf(x)
print(p)

```

```python
# x = np.array([0.3, 0.1])
mu = np.array([0.3, 0.7])
sigma = np.array([0.1, 0.2])

# make sigma into a diagonal matrix
sigma = np.diag(sigma)

# calculate the pdf
rv = multivariate_normal(mean=mu, cov=sigma)

x, y = np.mgrid[0:1:.001, 0:1:.001]
pos = np.dstack((x, y))
fig2 = plt.figure()
ax2 = fig2.add_subplot(111, projection='3d')
# add 3d contours
ax2.contourf(x, y, rv.pdf(pos))

```
